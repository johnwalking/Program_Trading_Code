```
import csv
bert_file_list = []
bert_file_list_split = []
with open('/content/drive/My Drive/Program_Trading_Code/bert_output.tsv') as csv_file:
  csv_reader = csv.reader(csv_file)
  for line in csv_reader:
    bert_file_list.append(line)
    line_list = line[0].split('\t')
    if line_list[0] == 'article_id': continue
    for i in [0,1,2]: line_list[i] = int(line_list[i])
    bert_file_list_split.append(line_list)
```

```
with open('/content/drive/My Drive/Program_Trading_Code/crf_output.tsv') as csv_file:
  csv_reader = csv.reader(csv_file)
  for line in csv_reader:
    line_list = line[0].split('\t')
    if line_list[0] == 'article_id': continue
    for i in [0,1,2]: line_list[i] = int(line_list[i])
    if line_list[4] in ['money']:
      if line in bert_file_list: continue
      else: bert_file_list_split.append(line_list)
print(bert_file_list_split[-5:])
bert_file_list_split.sort()
```

```
df = pd.DataFrame(bert_file_list_split, columns=['article_id', 'start_position', 'end_position', 'entity_text', 'entity_type'])
df.to_csv('result.tsv', sep ='\t', index=False)
```
